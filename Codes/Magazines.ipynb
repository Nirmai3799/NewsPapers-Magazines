```python
def getData(row):
    data = {
        'Name':str(row['పేరు']),
        'Editor_in_chief':str(row['ఎడిటర్-ఇన్-చీఫ్ ']),
        'Editor':str(row['ఎడిటర్']),
        'Formereditors':str(row['మాజీ ఎడిటర్స్ ']),
        'Based_in':str(row['ఆధారంగా']),
        'Publisher':str(row['ప్రచురణకర్త']),
        'Categories':str(row['కేటగిరీలు']),
        'Frequency':str(row['తరచుదనం']),
        # 'Firstissue':"Nan",
        'Company':str(row['కంపెనీ']),
        'Country':str(row['దేశం']),
        'Language':str(row['భాష']),
        'Founder':str(row['స్థాపకుడు']),
        'Firstissue':str(row['మొదటి ప్రచురణ ']),
        'YearFounded':str(row['స్థాపించబడిన సంవత్సరం']),
        'Circulation':str(row['సర్క్యులేషన్']),
        'ISSN':str(row['ISSN']),
        
    }
    return data
```

```python
# def main():
#     with open("Magzines.j2",encoding="utf8") as f:
#         S_template = Template(f.read())
#     for i in range(1,14):
#         row = df.iloc[i]
#         # title = str(row['Name']).strip()
#         text = S_template.render(getData(row))
#         # print('\n', i, title,'\n')
#         print(text)

def main():
	file_loader = FileSystemLoader('./')
	env = Environment(loader=file_loader)
	template = env.get_template('Magzines.j2')

	# Load the data (.pkl)
	moviesDF =pickle.load(open('./123.pkl', 'rb'))
	
	# Test XML generation 
	# ids = ['tt0111161','tt0252487','tt0068646','tt0050083','tt0093603','tt0468569','tt0252488','tt0167260','tt0110912','tt10888594','tt0417056','tt0096870','tt0060666','tt0421051','tt0808240','tt5988370','tt4009460','tt7886848','tt6038600','tt7221896']

	#remove this to generate articles for all movies
	ids = moviesDF.IMDbID.tolist()
	ids =ids[200:300] 

	# Initiate the file object
	fobj = open('Magazines_xml.xml', 'w')
	fobj.write(tewiki+'\n')

	

	# Give the page_id from which you want to generate the articles in
	initial_page_id = 500000
	rh=1

	# Loop to grab all data from the .pkl and generate articles using the template
	for i, movieId in enumerate(ids):
		if(rh==1):
			rh+=1
			continue
		# row = moviesDF.loc[moviesDF['IMDbID']==movieId]
		row = df.iloc[i]
		title = str(row['పేరు'])
		# if(str(row['స్థాపకుడు'])=="nan"):
		# 	continue

		text = template.render(getData(row))

		writePage(initial_page_id,title,text,fobj)
		initial_page_id += 1
		print(text, '\n')
		print(rh,end=" ")
		if(rh==50):
			break
		
		rh+=1

	fobj.write('</mediawiki>')
	fobj.close()
```
