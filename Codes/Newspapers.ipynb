```python
import pandas as pd
```

```python
from jinja2 import Template
```

```python
df = pd.read_csv("newscsv.csv")
print(df)
```

```python
def getData(row):
    data = {
        'Name':str(row['Title']),
        'Registration_no':str(row['Registration No']),
        'Based_in':str(row['State']),
        'Language':str(row['Language']),
        'Periodicity':str(row['Periodicity']),
        'Publication_city':str(row['Publication City']),
        'Owner_name':str(row['Owner Name']),
        
        
    }
    return data
```

```python
def main():
    with open("Newspapers.j2",encoding="utf8") as f:
        S_template = Template(f.read())
    for i in range(1,1400):
        row = df.iloc[i]
        # title = str(row['Name']).strip()
        text = S_template.render(getData(row))
        # print('\n', i, title,'\n')
        print(text)
```

```python
main()
```
